## Standardizing column names in order to maintain consistency and ease of merge

```python
def standardize_columns(df):
    df.columns = [col.title().replace(" ", "_") for col in df.columns]
standardize_columns(df1)
standardize_columns(df2)
standardize_columns(df3)
```

```python
merged_df = df1.merge(df3, on='Car_Id')
```

```python
merged_df.head()
```

```python
final = merged_df.merge(df2, on='Cust_Id')
```
